```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Load Dataset
dataset = pd.read_csv('/content/Salary_Data.csv')
```

```python
plt.scatter(dataset['YearsExperience'],dataset['Salary'])
plt.xlabel('Year of Experience')
plt.ylabel('salary')
plt.show()
```

Splitting data into Train Data and Test Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(dataset['YearsExperience'],dataset['Salary'],test_size = 0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train.values.reshape(-1,1),Y_train.values.reshape(-1,1))
Y_pred = regressor.predict(X_test.values.reshape(-1,1))
print(Y_pred)
```

Plotting the Graph

```python
plt.scatter(X_train,Y_train,color='Red')
plt.plot(X_train,regressor.predict(X_train.values.reshape(-1,1)),color ='blue')
plt.title('Salary v/s Experence(Trainig Set)')
plt.xlabel('Year oF Experience')
plt.ylabel('Salary')
plt.show()
```

Coefficient of Determination (r2 Score)

```python
from sklearn.metrics import r2_score
print('r2 Score : ', r2_score(Y_test, Y_pred))
```
